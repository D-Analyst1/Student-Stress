```python
## Rename Columns
rename_dict = {
    "Have you recently experienced stress in your life?": "stress_recent",
    "Have you noticed a rapid heartbeat or palpitations?": "heartbeat",
    "Have you been dealing with anxiety or tension recently?": "anxiety",
    "Do you face any sleep problems or difficulties falling asleep?": "sleep_problems",
    "Have you been getting headaches more often than usual?": "headaches",
    "Do you get irritated easily?": "irritability",
    "Do you have trouble concentrating on your academic tasks?": "concentration",
    "Have you been feeling sadness or low mood?": "sadness",
    "Have you been experiencing any illness or health issues?": "illness",
    "Do you often feel lonely or isolated?": "loneliness",
    "Do you feel overwhelmed with your academic workload?": "academic_workload",
    "Are you in competition with your peers, and does it affect you?": "peer_competition",
    "Do you find that your relationship often causes you stress?": "relationship_stress",
    "Are you facing any difficulties with your professors or instructors?": "professor_issues",
    "Is your working environment unpleasant or stressful?": "work_env",
    "Do you struggle to find time for relaxation and leisure activities?": "leisure_time",
    "Is your hostel or home environment causing you difficulties?": "home_env",
    "Do you lack confidence in your academic performance?": "confidence_academic",
    "Do you lack confidence in your choice of academic subjects?": "confidence_subjects",
    "Academic and extracurricular activities conflicting for you?": "academic_conflict",
    "Do you attend classes regularly?": "class_attendance",
    "Have you gained/lost weight?": "weight_change",
    "Which type of stress do you primarily experience?": "stress_type"
}

stress2= Stress.rename(columns=rename_dict)
stress2
```

```python
stress2
```

```python
## rename Gender column
stress2["Gender"] = stress2["Gender"].replace({1: "Male", 0: "Female"})
```

```python
stress2["Gender"].unique()
```

```python
stress2
```

```python
stress2['Age'].unique()
```

```python
stress3 = stress2[stress2["Age"] != 100]
stress3
```

```python
stress3["stress_type"] = stress3["stress_type"].replace({"Eustress (Positive Stress) - Stress that motivates and enhances performance.": "Postive Stress", 
                                                         "No Stress - Currently experiencing minimal to no stress.": "No Stress",
                                                        'Distress (Negative Stress) - Stress that causes anxiety and impairs well-being.': "Negative Stress"})
```

```python
stress3['stress_type'].unique()
```

```python
## Categorize Age Column
def age_category(age):
    if age <= 18:
        return "Teen"
    elif age <= 25:
        return "Young Adult"
    elif age <= 40:
        return "Adult"
    else:
        return "Older Adult"

stress3["AgeGroup"] = stress3["Age"].apply(age_category)
stress3

```

## students’ demographics and stress factors.

```python
# Distribution of Age

plt.hist(stress3["Age"], bins=10, edgecolor="black")
plt.title("Age Distribution of Students")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
## Distribution of Age Groups
stress3["AgeGroup"].value_counts().plot(kind="bar", color=['Gold', 'Brown','Black','Purple'], edgecolor="black")
plt.title("Age Group Distribution")
plt.xlabel("Age Group")
plt.ylabel("Number of Students")
plt.show()
```

```python
## Gender Distribution
stress3["Gender"].value_counts().plot(kind="bar", color=['Gold', 'Brown'])
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Number of Students")
plt.show()
```

```python
## Stress Type Distribution
stress3["stress_type"].value_counts().plot(kind="bar", color=['Gold', 'Brown','Black'], edgecolor="black")
plt.title("Types of Stress Experienced")
plt.xlabel("Stress Type")
plt.ylabel("Number of Students")
plt.show()
```

## Bivariate Analysis

```python
## Stress Type vs Gender

sns.countplot(data=stress3, x="stress_type", hue="Gender")
plt.title("Stress Type by Gender")
plt.xlabel("Stress Type")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
## Stress Type vs Age Group

sns.displot(stress3, x="stress_type", col="AgeGroup")
```

```python
## Key Stress Factors vs Gender (example: Sleep Problems)
sns.countplot(data=stress3, x="sleep_problems", hue="Gender", color='darkgreen')
plt.title("Sleep Problems by Gender")
plt.xlabel("Sleep Problems")
plt.ylabel("Count")
plt.show()
```

```python
numeric_df = stress3.select_dtypes(include=["number"])
```

```python
## Correlation Between Stress Factors
plt.figure(figsize=(16,7))
sns.heatmap(numeric_df .corr(), annot=True, fmt=".1f", cmap="crest")
plt.title("Correlation Between Stress Factors")
plt.show()
```

## Insights & Interpretation

```python
#  4 = Correlation between Stress Factors

  # The heatmap suggests that student stress is a network of interconnected challenges rather than isolated issues. Academic pressure often sets off a chain reaction: workload → anxiety → poor sleep → physical symptoms → emotional struggles. 
  #Once the cycle starts, it reinforces itself, making stress management more complex.
```

```python
# 5 = sleep problems by Gender 

   # Sleep problems are not a minor issue but a core stress indicator among students. They reflect and reinforce both physical and emotional stress factors. 
  # The gender gap suggests that female students experience and possibly internalize stress differently, which translates into more frequent and severe sleep challenges. 
 # Addressing student stress effectively will require holistic strategies that combine academic support, mental health resources, and sleep-focused interventions.
```
